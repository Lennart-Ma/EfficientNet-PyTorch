```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
perc_data = [100, 50, 25, 10, 5, 2]
fid_score_STYLEGAN2ADA = [43.45, 46.9, 54.37, 47.82, 53.708, 79.95]
fid_score_STYLEGAN2 = [69.103, 75.65, 100.68, 70.5, 70.5, 70.5]

fig = plt.figure()
ax=fig.add_subplot()
ax2=fig.add_subplot(frame_on=False)
ax.plot(perc_data, fid_score_STYLEGAN2ADA, marker='o', color="C0")
ax2.plot(perc_data,fid_score_STYLEGAN2, marker='o', color="C1")
ax.set(xlabel="% training data", ylabel="AUC")
ax.set_ylim([20, 110])
ax2.set_ylim([20, 110])
ax.legend(["StyleGAN2 ADA"], loc=(0.6, 0.9))
ax2.legend(["StyleGAN2"], loc=(0.6, 0.8))
ax.grid()
ax2.grid()
ax.xaxis.set_ticks([2, 5, 10, 25, 50, 100])
plt.show()
```

```python
perc_data = [100, 50, 25, 10, 5, 2]
auc_nofake= [0.962, 0.932, 0.718, 0.718, 0.7033, 0.6208]
auc_fake= [0.976, 0.895, 0.874, 0.7334, 0.6974, 0.6499]

fig = plt.figure()
ax=fig.add_subplot(111, label="1")
ax2=fig.add_subplot(111, label="2", frame_on=False)
ax.plot(perc_data, auc_nofake, marker='o', color="C0")
ax2.plot(perc_data,auc_fake, marker='o', color="C1")
ax.set(xlabel="% training data", ylabel="AUC")
ax.set_ylim([0.5,1])
ax2.set_ylim([0.5,1])
ax.legend(["without fake data"], loc=(0.6, 0.7))
ax2.legend(["with fake data"], loc=(0.6, 0.6))
ax.grid()
ax2.grid()
ax.xaxis.set_ticks([2, 5, 10, 25, 50, 100])
plt.show()

```
